# 7. Operatori

## Aritmētiskie operatori
- `+` saskaitīšana
- `-` atņemšana
- `*` reizināšana
- `/` dalīšana
- `//` veselā dalīšana
- `%` atlikums
- `**` pakāpe

## Salīdzināšanas operatori
- `==`
- `!=`
- `>`
- `<`
- `>=`
- `<=`

## Loģiskie operatori
- `and`
- `or`
- `not`

```python
a = 10
b = 3

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)
```

```python
a = 10
b = 3

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= 10)
print(b <= 2)
```

```python
ir_pilngadigs = True
ir_registrets = False

print(ir_pilngadigs and ir_registrets)
print(ir_pilngadigs or ir_registrets)
print(not ir_registrets)
```
